```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Range for values (true and predicted)
y = np.linspace(0, 1, 100)
y_hat = np.linspace(0, 1, 100)
Y, Y_hat = np.meshgrid(y, y_hat)
```

```python
epsilon = 1e-15

# MSE
MSE = (Y - Y_hat) ** 2

# BCE
BCE = - (Y * np.log(Y_hat + 1e-15) + (1 - Y) * np.log(1 - Y_hat + epsilon))
```

```python
# Plot MSE
fig = plt.figure(figsize=(14, 6))

ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(Y, Y_hat, MSE, cmap='viridis')
ax.set_xlabel('True Value (y)')
ax.set_ylabel('Predicted Value (y_hat)')
ax.set_zlabel('MSE')
ax.set_title('Mean Squared Error (MSE)')

# Plot BCE
ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(Y, Y_hat, BCE, cmap='viridis')
ax.set_xlabel('True Value (y)')
ax.set_ylabel('Predicted Value (y_hat)')
ax.set_zlabel('BCE')
ax.set_title('Binary Cross Entropy (BCE)')

plt.tight_layout()
plt.show()
```
